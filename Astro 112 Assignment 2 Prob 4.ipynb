```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
from astropy.io import ascii
```

```python
profile05 = ascii.read('/Users/sarahbowman/Desktop/mesa-r10398/tutorial05/LOGS/profile17.data',header_start=4,data_start=5)
profile1 = ascii.read('/Users/sarahbowman/Desktop/mesa-r10398/tutorial1/LOGS/profile17.data',header_start=4,data_start=5)
profile8 = ascii.read('/Users/sarahbowman/Desktop/mesa-r10398/tutorial8/LOGS/profile20.data',header_start=4,data_start=5)
profile128 = ascii.read('/Users/sarahbowman/Desktop/mesa-r10398/tutorial128/LOGS/profile8.data',header_start=4,data_start=5)

names = profile05.colnames
print names

dlnT05 = profile05['logT'][1:]-profile05['logT'][0:-1]
dlnP05 = profile05['logP'][1:]-profile05['logP'][0:-1]

dlnT1 = profile1['logT'][1:]-profile1['logT'][0:-1]
dlnP1 = profile1['logP'][1:]-profile1['logP'][0:-1]

dlnT8 = profile8['logT'][1:]-profile8['logT'][0:-1]
dlnP8 = profile8['logP'][1:]-profile8['logP'][0:-1]

dlnT128 = profile128['logT'][1:]-profile128['logT'][0:-1]
dlnP128 = profile128['logP'][1:]-profile128['logP'][0:-1]

plt.figure(figsize=(10,6))
plt.plot(np.e**(profile05['logR'][1:])/np.e**(np.max(profile05['logR'])), dlnT05/dlnP05 ,color='indigo',label='0.5 $M_{\odot}$ star')
plt.plot(np.e**(profile1['logR'][1:])/np.e**(np.max(profile1['logR'])), dlnT1/dlnP1 ,label='1 $M_{\odot}$ star')
plt.plot(np.e**(profile8['logR'][1:])/np.e**(np.max(profile8['logR'])), dlnT8/dlnP8 ,color='green',label='8 $M_{\odot}$ star')
plt.plot(np.e**(profile128['logR'][1:])/np.e**(np.max(profile128['logR'])), dlnT128/dlnP128 ,color='deepskyblue',label='128 $M_{\odot}$ star')

plt.title('dlog(T)/dlog(P) as a function of R/$R_*$ (until star reaches Main Sequence)')
plt.legend(loc='best')
plt.semilogx()
plt.xlabel('R/$R_*$')
plt.xlim(-0.1,1.1)
plt.ylabel('dlog(T)/dlog(P)')
plt.grid()
plt.axhspan(2/5.,1/4., alpha=0.5, color='fuchsia')

plt.savefig('hw2p4plot.pdf')
```
