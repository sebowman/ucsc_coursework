```python
from numpy import *
%matplotlib inline
from matplotlib.pyplot import *
from astropy.io import ascii

history05 = ascii.read('/Users/sarahbowman/Desktop/history0.5.data.txt',header_start=4,data_start=5)
history1 = ascii.read('/Users/sarahbowman/Desktop/history1.data.txt',header_start=4,data_start=5)
history8 = ascii.read('/Users/sarahbowman/Desktop/history8.data.txt',header_start=4,data_start=5)
history128 = ascii.read('/Users/sarahbowman/Desktop/history128.data.txt',header_start=4,data_start=5)

figure(figsize=(10,6))
plot(history05['star_age'], history05['log_R'],color='indigo',label='0.5 $M_{\odot}$ star')
plot(history1['star_age'], history1['log_R'],label='1 $M_{\odot}$ star')
plot(history8['star_age'], history8['log_R'],color='green',label='8 $M_{\odot}$ star')
plot(history128['star_age'], history128['log_R'],color='deepskyblue',label='128 $M_{\odot}$ star')
title('Radius as a function of time (until star reaches Main Sequence)')
legend()
semilogx()
xlabel('age [yr]')
ylabel('log(Radius) [solar]')
grid()

savefig('rvt.pdf')
```

```python
history1 = ascii.read('/Users/sarahbowman/Desktop/history1.data.txt',header_start=4,data_start=5)

history = history1

names = history.colnames

figure(figsize=(10,6))
plot(history['star_age'], history['log_R'])
semilogx()
title('Radius as a function of time for a 1 Solar Mass Star')
xlabel('age [yr]')
ylabel('log(Radius) [solar]')
grid()

savefig('1sm.pdf')
```
