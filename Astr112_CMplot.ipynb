```python
br = df.bp_rp
p = df.parallax
g = df.phot_g_mean_mag
ag = df.a_g_val
ebr = df.e_bp_min_rp_val #etc
temp = df.teff_val
```

```python
type(g)
```

```python
color = br-ebr
mag = g - (5*np.log10(1./p*1000)-5)-ag
```

```python
plt.figure(figsize=(5,5))
plt.scatter(color,mag, marker ='o',c=temp,cmap='jet_r',edgecolors='none')
#plt.plot(color,mag,'.')

plt.ylim(9,-12)
plt.xlabel('BP - RP color')
plt.ylabel('Absolute mag')
plt.colorbar(label='Eff T (K)')
```
