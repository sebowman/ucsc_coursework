```python
import pandas as pd
df = pd.read_csv('1526957476347O-result.csv')
df2 = pd.read_csv('1526963339973O-result.csv')
```

```python
br = df.bp_rp
p = df.parallax
p2 = df2.parallax
g = df.phot_g_mean_mag
#ag = df.a_g_val
temp = df.teff_val
tp = df2.spectype
```

```python
type(g)
```

```python
da = df2[tp=='DA']
db = df2[tp=='DB']
dq = df2[tp=='DQ']
bra = da.bp_rp
pa = da.parallax
ga = da.phot_g_mean_mag
brb = db.bp_rp
pb = db.parallax
gb = db.phot_g_mean_mag
brq = dq.bp_rp
pq = dq.parallax
gq = dq.phot_g_mean_mag
```

```python
color = br
mag = g - (5*np.log10(1./p*1000)-5)
colora = bra
maga = ga - (5*np.log10(1./pa*1000)-5)
colorb = brb
magb = gb - (5*np.log10(1./pb*1000)-5)
colorq = brq
magq = gq - (5*np.log10(1./pq*1000)-5)
```

```python
plt.figure(figsize=(5,5))
#plt.scatter(color,mag, marker ='.', s = .05,c=temp,cmap='jet_r',edgecolors='none')
plt.scatter(color,mag, marker ='.', s = .05)
plt.scatter(colora,maga, marker ='.', s = .05, color= 'blue')
plt.scatter(colorb,magb, marker ='.', s = .05, color= 'green')
plt.scatter(colorq,magq, marker ='.', s = .05, color= 'red')
#plt.plot(color,mag,'.')

plt.ylim(16,-16)
plt.xlabel('BP - RP color')
plt.ylabel('Absolute mag')
#plt.colorbar(label='Eff T (K)')
```

```python
plt.figure(figsize=(5,5))
#plt.scatter(color,mag, marker ='.', s = .05,c=temp,cmap='jet_r',edgecolors='none')
plt.scatter(color,mag, marker ='.', s = .05)
plt.scatter(colora,maga, marker ='.', s = .1, color= 'blue', label = 'DA white dwarfs')
plt.scatter(colorb,magb, marker ='.', s = .3, color= 'green', label = 'DB white dwarfs')
plt.scatter(colorq,magq, marker ='.', s = .3, color= 'red', label = 'DQ white dwarfs')
#plt.plot(color,mag,'.')

plt.legend()
plt.ylim(16,10)
plt.xlim(-1,2)
plt.xlabel('BP - RP color')
plt.ylabel('Absolute mag')
plt.title('Gaia HR Diagram: DA, DB, and DQ spectral classes')
#plt.colorbar(label='Eff T (K)')
```

```python
from scipy.stats import kde

index = np.logical_not(np.logical_or(np.isnan(mag), np.isnan(color)))
 
# create data
x = color[index][color<2][mag>10]
y = mag[index][color<2][mag>10]
 
# Evaluate a gaussian kde on a regular grid of nbins x nbins over data extents
nbins=500
k = kde.gaussian_kde([x,y])
xi, yi = np.mgrid[x.min():x.max():nbins*1j, y.min():y.max():nbins*1j]
zi = k(np.vstack([xi.flatten(), yi.flatten()]))
 
# Change color palette
plt.figure()

plt.pcolormesh(xi, yi, zi.reshape(xi.shape), cmap=plt.cm.Greens_r)
plt.ylim(16,10)
plt.show()
```
